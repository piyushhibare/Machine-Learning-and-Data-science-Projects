# Linear regression

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
X = data[['MMIN','MMAX','PRP','ERP']]
```

```python
y = data['CACH']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.33,random_state=42)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
pred = reg.predict(X)
```

```python
pred
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
rmse = mean_squared_error(y, pred)
```

```python
rmse
```

# Linear regression using Ridge

```python
reg1 = linear_model.Ridge (alpha = .5)
```

```python
reg1.fit(X_train, y_train)
```

```python
pred1 = reg1.predict(X)
```

```python
pred1
```

```python
reg1.coef_
```

```python
reg1.intercept_
```

# Linear regression using Lasso

```python
reg2 = linear_model.Lasso(alpha = 0.1)
```

```python
reg2.fit(X_train,y_train)
```

```python
pred = reg2.predict(X_test)
```
