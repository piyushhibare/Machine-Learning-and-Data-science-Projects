```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/bridges/bridges.data.version1')
```

```python
data.columns = ['IDENTIF','RIVER','LOCATION','ERECTED','PURPOSE','LENGTH','LANES','CLEAR-G','T-OR-D','MATERIAL','SPAN','REL-L','TYPE']
```

```python
data
```

```python
data['T-OR-D'].count()
```

```python
#from sklearn.preprocessing import OneHotEncoder
```

```python
#enc=OneHotEncoder(sparse=False)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
X = data[['RIVER','PURPOSE','CLEAR-G','T-OR-D','MATERIAL','SPAN','REL-L']]
```

```python
y = data['TYPE']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33, random_state = 42)
```

# Convert categorical data into numerical data using Label Encoder

```python
for col in X_test.columns.values:
    if X_test[col].dtypes=='object':
               data=X_train[col].append(X_test[col])
               le.fit(data.values)
               X_train[col]=le.transform(X_train[col])
               X_test[col]=le.transform(X_test[col])
```

```python
X_test
```

```python
X_train
```

# Classification using support vector machine

```python
#import support vector machine
from sklearn.svm import SVC
#import dependecies for find accuracy
from sklearn.metrics import accuracy_score
```

```python
# create support vector classifier
clf = SVC()
```

```python
# fit svm
clf.fit(X_train,y_train)
```

```python
# find prediction 
pred = clf.predict(X_test)
```

```python
pred
```

```python
accuracy_score(y_test, pred)
```

# Classification using naive bayes 

```python
# import naive bayes
from sklearn.naive_bayes import GaussianNB
```

```python
clf1 = GaussianNB()
```

```python
clf1.fit(X_train,y_train)
```

```python
pred = clf1.predict(X_test)
```

```python
pred
```

```python
accuracy_score(y_test, pred)
```

# Classification using decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf2 = DecisionTreeClassifier(random_state=0)
```

```python
clf2.fit(X_train,y_train)
```

```python
pred = clf2.predict(X_test)
```

```python
pred
```

```python
cross_val_score(clf2, X_test, pred, cv=10)
```
