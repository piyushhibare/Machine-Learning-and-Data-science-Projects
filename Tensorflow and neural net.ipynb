# Simple regression example

```python
X_data = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
y_data = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X_data,y_data,'.')
```

```python
np.random.rand(2)
```

```python
np.random.seed(100)
```

```python
np.random.rand(2)
```

```python
m = tf.Variable(0.54)
b = tf.Variable(0.27)
```

```python
error = 0

for x,y in zip(X_data,y_data):
  
  y_hat = m*x + b
  
  error += (y-y_hat)**2
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
```

```python
train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    training_steps = 100
    
    for i in range(training_steps):
        
        sess.run(train)
        
    final_slope, final_intercept = sess.run([m,b])
```

```python
x_test = np.linspace(1,10,10)
```

```python
# y = mx+b
y_pred_plot = final_slope*x_test + final_intercept 
```

```python
plt.plot(x_test,y_pred_plot,'r')
plt.plot(X_data,y_data,'*')
```

```python
y_pred_plot
```

# Realistics regression

```python
import pandas as pd
```

```python
X_data2 = np.linspace(0.0,10.0,1000000)
```

```python
noise = np.random.random([len(X_data2)])
```

```python
noise
```

```python
y_true = (0.5 * X_data2) + 5 + noise
```

```python
X_df = pd.DataFrame(data=X_data2,columns = ['X_data'])
```

```python
Y_df = pd.DataFrame(data=y_true,columns = ['y'])
```

```python
X_df.head()
```

```python
Y_df.head()
```

```python
my_data = pd.concat([X_df,Y_df],axis=1)
```

```python
my_data.head()
```

```python
my_data.sample(1000).plot(kind='scatter',x='X_data',y='y')
```

```python
batch_size = 8
```

```python
m = tf.Variable(5.2)
b = tf.Variable(0.9)
```

```python
xph = tf.placeholder(tf.float32,[batch_size])
yph = tf.placeholder(tf.float32,[batch_size])
```

```python
y_model = m * xph + b
```

```python
error = tf.reduce_sum(tf.square(yph-y_model))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
```

```python
train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    batches = 10000
    
    for i in range(batches):
        
        rand_ind = np.random.randint(len(X_data2),size=batch_size)
        
        feed = {xph:X_data2[rand_ind],yph:y_true[rand_ind]}
        
        sess.run(train,feed_dict = feed)
        
    model_m, model_b = sess.run([m,b])
```

```python
y_hat = X_data2 * model_m + model_b
```

```python
my_data.sample(250).plot(kind='scatter',x='X_data',y='y')
plt.plot(X_data2,y_hat,'r')
```
