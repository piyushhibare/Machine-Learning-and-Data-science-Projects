```python
ax = plt.subplots(figsize=(10, 10))
g = sns.heatmap(data.corr(), square=True, annot=True);
```

```python
g = sns.pairplot(data,
                 vars=['x', 'y', 'z', 'carat'],
                 size=4,
                 diag_kind='kde',
                 diag_kws=dict(shade=True));
```

```python
g = sns.lmplot('carat', 'price', data=data, fit_reg=False, scatter=True, size=16);
```

```python
sns.set_style("whitegrid")
ax = sns.violinplot(x="carat", y="price", data=data)
```

```python
fig, ax = plt.subplots(figsize=(12,6)) 

for cut in data['cut'].unique():
    g = sns.kdeplot(data[data['cut'] == cut]['depth'], shade=True, ax=ax);

plt.title('Depth (%) KDE for different cuts');
plt.xlabel('Depth (%)');
plt.ylabel('KDE');
plt.legend(data['cut'].unique());

f, ax = plt.subplots(figsize=(12,8))
g = sns.violinplot(x='cut', y="depth", data=data);
ax.set_title("Cut vs depth (%)");
g.set_xlabel('');
g.set_ylabel('Depth (%)');
```
